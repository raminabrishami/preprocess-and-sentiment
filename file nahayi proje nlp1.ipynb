# در این قسمت تابع های پری پراسس و ان گرام و ... اورده شده 

```python
def preprocess(matn_vorodi):

    global lemmatized_words
    global filtered_list
    global fixd_lemmatize_words
    global text
    global token
    global stop_words

    text = matn_vorodi
    
    text=text.lower()
    
    # هشتگ ها و @ ها را حذف کردیم
    
    spl=text.split()

    dell=[]

    for i in spl:
        if '@' in i:
            dell.append(i)
        elif '#' in i:
            dell.append(i)

    for i in dell:
        spl.remove(i)

    text=' '.join(spl)
    
    # 
     
    text=re.sub(r"/^(?:([A-Za-z]+):)?(\/{0,3})([0-9.\-A-Za-z]+)(?::(\d+))?(?:\/([^?#]*))?(?:\?([^#]*))?(?:#(.*))?$/;",r' ',text)
                
    token=word_tokenize(text)

    stop_words = set(stopwords.words("english"))
    
    for i in remov:
        stop_words.remove(i)
    stop_words.add(')')
    stop_words.add(':')
    stop_words.add('@')
    stop_words.add('#')

    filtered_list = []
    
    for word in token:
        if word.casefold() not in stop_words:
             filtered_list.append(word)

    stemmer = PorterStemmer()
    
    stemmer_words = [stemmer.stem(word) for word in filtered_list]
    
    lemmatizer = WordNetLemmatizer()

    lemmatized_words=[lemmatizer.lemmatize(word,'v') for word in filtered_list]

  # fixd_lemmatize_words=' '.join(stemmer_words)
    
    fixd_lemmatize_words=' '.join(lemmatized_words)

    return (fixd_lemmatize_words)


def bag_of_words():

    global bag_of_text
    global bag
    global bagg

    bag_of_text = {}

    for i in unic_textz:
        bag_of_text[i] =0


    for word in lemmatized_words:
        if word in unic_textz:
            bag_of_text[word] += 1


    bag=bag_of_text.values()

    bagg.append(list(bag))


def N_gram(n):

    global Count_data_train
    global Count_data_test
    global CountVec

    CountVec = CountVectorizer(ngram_range=(n,n))

    Count_data_train = CountVec.fit_transform(tweets_train)
    
    Count_data_test = CountVec.transform(tweets_test)


# برای روش cv 

def N_gram_cv(n):

    global Count_data_train

    CountVec = CountVectorizer(ngram_range=(n,n))

    Count_data_train = CountVec.fit_transform(tweet)
    
    
# ترکیب روش انگرام و تی اف ای دی اف

def N_tf(n):
    
    global Count_data_train
    global Count_data_test
    global tf_idf_vec_smooth
    
    tf_idf_vec_smooth = TfidfVectorizer(use_idf=True,  
                            smooth_idf=True,  
                            ngram_range=(n,n),stop_words='english')


    Count_data_train = tf_idf_vec_smooth.fit_transform(tweets_train)
    
    Count_data_test= tf_idf_vec_smooth.transform(tweets_test)
    
    
# برای روش cv 
    
def N_tf_cv(n):
    
    global Count_data_train

    tf_idf_vec_smooth = TfidfVectorizer(use_idf=True,  
                            smooth_idf=True,  
                            ngram_range=(n,n),stop_words='english')


    Count_data_train = tf_idf_vec_smooth.fit_transform(tweet)
    
    

def tf_idf():

    global Count_data_train
    global Count_data_test
    global tfidf

    tfidf = TfidfVectorizer()

    Count_data_train = tfidf.fit_transform(tweets_train)
    
    Count_data_test = tfidf.transform(tweets_test)

    
    
# برای روش cv     
    
def tf_idf_cv():

    global Count_data_train

    tfidf = TfidfVectorizer()

    Count_data_train = tfidf.fit_transform(tweet)
    
    
    
#روش شمارش تکرار کلمه ها در چملات مثبت و منفی 
    
def shomar(tweets,labels):
    f={}
    for i,j in zip(tweets,labels):
        i=i.split()
        for word in i:
            p=(word,j)
            f[p]=f.get(p,0)+1
    return f


def get_features_shomar(tweet,f):

    z=np.zeros((1,3))
    z[0,0]=1
    i=tweet.split()
    for j in i:
        z[0,1]+=f.get((j,1),0)
        z[0,2]+=f.get((j,0),0)
        #print(z)
        
    assert(z.shape==(1,3))
    return z
            

```
